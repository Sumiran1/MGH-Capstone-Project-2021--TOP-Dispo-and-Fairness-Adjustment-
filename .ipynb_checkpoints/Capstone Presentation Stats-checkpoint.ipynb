```python
pen = pd.concat([train_pen, test_pen])
blunt = pd.concat([train_blunt, test_blunt])
```

```python
pen.columns
```

```python
from scipy.stats import pearsonr

column_values = [column for column in pen[['age', 'sbp1', 'pulse1', 'oxysat1', 'temp1', 'rr1', 'gcstot1']].columns.tolist() ]


pen['Correlation_coefficent'], pen['P-value'] = zip(*pen.T.apply(lambda x: pearsonr(x[column_values ],x[column_values ])))
df_result = pen[['Correlation_coefficent','P-value']]
df_result
```

```python
from scipy.stats import pearsonr

column_values = [column for column in blunt[['age', 'sbp1', 'pulse1', 'oxysat1', 'temp1', 'rr1', 'gcstot1']].columns.tolist() ]


blunt['Correlation_coefficent'], blunt['P-value'] = zip(*blunt.T.apply(lambda x: pearsonr(x[column_values ],x[column_values ])))
df_result = blunt[['Correlation_coefficent','P-value']]
df_result
```

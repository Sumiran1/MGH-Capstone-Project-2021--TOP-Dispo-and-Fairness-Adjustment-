```python
results = pd.read_csv("fairness_oct_split_results_depth10.csv")
results['train_diff'] = results.aucs_updated_train - results.aucs_original_train
results['test_diff'] = results.aucs_updated_test - results.aucs_original_test
results
```

```python
results_stats = results.describe()
results_stats.loc['Standard error over mean'] = results.sem().to_numpy()
results_stats = results_stats.loc[['mean','std','Standard error over mean']]
results_stats['Model'] = ['Optimal Classification Trees'] * 3
results_stats.to_csv("C:/Users/15125/Desktop/CAPSTONE PROJECT/CLEANED DATA/FAIRNESS MODEL/aucs_oct.csv")
```

```python
results_change_no_race = pd.read_csv("fairness_oct_split_change_of_label_results_depth5_norace.csv")
results_change_no_race.columns = ['overall_acc_train', 'p/n_acc_train', 'overall_acc_test', 'p/n_acc_test']
results_change_no_race['overall_acc_diff'] = np.abs(results_change_no_race.overall_acc_train - results_change_no_race.overall_acc_test)
results_change_no_race['p/n_acc_diff'] = np.abs(results_change_no_race['p/n_acc_train'] - results_change_no_race['p/n_acc_test'])
results_change_no_race = results_change_no_race.describe()
results_change_no_race.loc['Standard error over mean'] = results_change_no_race.sem().to_numpy()
results_change_no_race
```

```python
results_change_race = pd.read_csv("fairness_oct_split_change_of_label_results_depth5_race.csv")
results_change_race.columns = ['overall_acc_train', 'p/n_acc_train', 'overall_acc_test', 'p/n_acc_test']
results_change_race['overall_acc_diff'] = np.abs(results_change_race.overall_acc_train - results_change_race.overall_acc_test)
results_change_race['p/n_acc_diff'] = np.abs(results_change_race['p/n_acc_train'] - results_change_race['p/n_acc_test'])
results_change_race = results_change_race.describe()
results_change_race.loc['Standard error over mean'] = results_change_race.sem().to_numpy()
results_change_race
```

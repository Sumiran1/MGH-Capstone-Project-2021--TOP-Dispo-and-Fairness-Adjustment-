```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.metrics import roc_curve, auc
```

# Morbidity and Mortality

```python
data_path = "../../../../../../Dropbox (MIT)/TQIP OCT MGH-MIT/folder_y_preds/01_02_2020/"

# Morbidity
y_pred_morbid_penetrating = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/y_pred_proba_seed=1___outcome=hosp_morbidity___minbucket=100___injury=penetrating.csv"
).iloc[:,1]
y_pred_morbid_blunt = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/y_pred_proba_seed=1___outcome=hosp_morbidity___minbucket=100___injury=blunt.csv"
).iloc[:,1]
y_test_morbid_penetrating = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/test_y_morbid_penetrating.csv"
).iloc[:,1]
y_test_morbid_blunt = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/test_y_morbid_blunt.csv"
).iloc[:,1]



#Mortality
y_pred_mortal_penetrating = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/y_pred_proba_seed=1___outcome=hosp_mortality___minbucket=100___injury=penetrating.csv"
).iloc[:,1]
y_pred_mortal_blunt = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/y_pred_proba_seed=1___outcome=hosp_mortality___minbucket=100___injury=blunt.csv"
).iloc[:,1]
y_test_mortal_penetrating = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/test_y_mortal_penetrating.csv"
).iloc[:,1]
y_test_mortal_blunt = pd.read_csv(
    data_path+"outputs_general_y_pred_proba_01_02_2020/test_y_mortal_blunt.csv"
).iloc[:,1]

print("Downloaded all the data.")
```

```python
# Morbidity
fpr_morbid_pen, tpr_morbid_pen, thresh_morbid_pen = roc_curve(
    y_test_morbid_penetrating,
    y_pred_morbid_penetrating
)
auc_morbid_pen = auc(fpr_morbid_pen, tpr_morbid_pen)

fpr_morbid_blunt, tpr_morbid_blunt, thresh_morbid_blunt = roc_curve(
    y_test_morbid_blunt,
    y_pred_morbid_blunt
)
auc_morbid_blunt = auc(fpr_morbid_blunt, tpr_morbid_blunt)

plt.figure(figsize=(12, 7))
plt.title(f'ROC Curve for Morbidity (Blunt and Penetrating)', fontsize=16)
plt.plot(fpr_morbid_pen, tpr_morbid_pen, 'b', label='Test AUC Penetrating = %0.3f' % auc_morbid_pen, color='g')
plt.plot(fpr_morbid_blunt, tpr_morbid_blunt, 'b', label='Test AUC Blunt = %0.3f' % auc_morbid_blunt, color='r')
plt.plot([0, 1], [0, 1], 'k--', label="Baseline AUC = 0.500")
plt.legend(loc='lower right', fontsize=12)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate', fontsize=12)
plt.xlabel('False Positive Rate', fontsize=12)
plt.savefig(f"{datetime.now().date()}_roc_curve_morbidity.png")
plt.show()
```

```python
# mortality
fpr_mortal_pen, tpr_mortal_pen, thresh_mortal_pen = roc_curve(
    y_test_mortal_penetrating,
    y_pred_mortal_penetrating
)
auc_mortal_pen = auc(fpr_mortal_pen, tpr_mortal_pen)

fpr_mortal_blunt, tpr_mortal_blunt, thresh_mortal_blunt = roc_curve(
    y_test_mortal_blunt,
    y_pred_mortal_blunt
)
auc_mortal_blunt = auc(fpr_mortal_blunt, tpr_mortal_blunt)

plt.figure(figsize=(12, 7))
plt.title(f'ROC Curve for mortality (Blunt and Penetrating)', fontsize=16)
plt.plot(fpr_mortal_pen, tpr_mortal_pen, 'b', label='Test AUC Penetrating = %0.3f' % auc_mortal_pen, color='g')
plt.plot(fpr_mortal_blunt, tpr_mortal_blunt, 'b', label='Test AUC Blunt = %0.3f' % auc_mortal_blunt, color='r')
plt.plot([0, 1], [0, 1], 'k--', label="Baseline AUC = 0.500")
plt.legend(loc='lower right', fontsize=12)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate', fontsize=12)
plt.xlabel('False Positive Rate', fontsize=12)
plt.savefig(f"{datetime.now().date()}_roc_curve_mortality.png")
plt.show()
```

## Blunt

```python
for comorb in [4, 5, 8, 12, 14, 19, 21, 25, 32]:
    (vars()[f"fpr_comorb_{comorb}_blunt"], 
     vars()[f"tpr_comorb_{comorb}_blunt"], 
     vars()[f"thresh_comorb_{comorb}_blunt"]
    ) = roc_curve(
        vars()[f"y_test_morbidity_{comorb}_blunt"],
        vars()[f"y_pred_morbidity_{comorb}_blunt"]
    )
    vars()[f"auc_comorb_{comorb}_blunt"] = auc(vars()[f"fpr_comorb_{comorb}_blunt"], 
                                               vars()[f"tpr_comorb_{comorb}_blunt"])
    print(f"Comorbidity {comorb} - Blunt")
    print(vars()[f"auc_comorb_{comorb}_blunt"])
```

```python
plt.figure(figsize=(12, 7))
plt.title(f'ROC Curve for co-morbidities (Blunt)', fontsize=16)
colors_dict = {
    4:'r', 5:'b', 8:'violet', 12:'g', 14:'darkorange', 19:'turquoise', 21:'firebrick', 25:'lime', 32:'purple'
}
for comorb in [4, 5, 8, 12, 14, 19, 21, 25, 32]:
    plt.plot(
        vars()[f"fpr_comorb_{comorb}_blunt"],
        vars()[f"tpr_comorb_{comorb}_blunt"],
        'b',
        label=f"Test AUC Co-Morbidity n°{comorb} = {round(vars()[f'auc_comorb_{comorb}_blunt'], 3)}",
        color=colors_dict[comorb]
    )
plt.plot([0, 1], [0, 1], 'k--', label="Baseline AUC = 0.500")
plt.legend(loc='lower right', fontsize=12)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate', fontsize=12)
plt.xlabel('False Positive Rate', fontsize=12)
plt.savefig(f"{datetime.now().date()}_roc_curve_comorb_blunt.png")
plt.show()
```

```python
import numpy as np
# Set data
df_auc_blunt = pd.DataFrame({
    'group': ['Test AUC','Fake'],
    'Acute Kidney Injury': [auc_comorb_4_blunt, 10],
    'ARDS': [auc_comorb_5_blunt, 10],
    'Cardiac Arrest CPR': [auc_comorb_8_blunt, 10],
    'Deep Surgical Site Infection': [auc_comorb_12_blunt, 10],
    'DVT': [auc_comorb_14_blunt, 10],
    'Organ Space Surgical Site Infection': [auc_comorb_19_blunt, 10],
    'Pulmonary Embolism': [auc_comorb_21_blunt, 10],
    'Unplanned Intubation': [auc_comorb_25_blunt, 10],
    'Severe Sepsis': [auc_comorb_32_blunt, 10],
})
 
# number of variable
categories=list(df_auc_blunt)[1:]
N = len(categories)
 
# We are going to plot the first line of the data frame.
# But we need to repeat the first value to close the circular graph:
values=df_auc_blunt.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories, color='grey', size=15)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([0.25, 0.5, 0.75], ["0.25","0.50","0.75"], color="grey", size=12)
plt.ylim(0,1)
plt.title(f"Test AUC for Comorbidities - Blunt", size=22, color="grey")
# Plot data
ax.plot(angles, values, linewidth=2, linestyle='solid')
# Fill area
ax.fill(angles, values, 'b', alpha=0.2)
plt.savefig(f"{datetime.now().date()}_radar_plot_comorb_blunt.png")
plt.show()
```

# Calibrations Curves

```python
from sklearn.calibration import calibration_curve
# Morbidity
fop_morbid_penetrating, mpv_morbid_penetrating = calibration_curve(
    y_test_morbid_penetrating, y_pred_morbid_penetrating, n_bins=10
)
fop_morbid_blunt, mpv_morbid_blunt = calibration_curve(
    y_test_morbid_blunt, y_pred_morbid_blunt, n_bins=10
)
# Mortality
fop_mortal_penetrating, mpv_mortal_penetrating = calibration_curve(
    y_test_mortal_penetrating, y_pred_mortal_penetrating, n_bins=10
)
fop_mortal_blunt, mpv_mortal_blunt = calibration_curve(
    y_test_mortal_blunt, y_pred_mortal_blunt, n_bins=10
)
plt.figure(figsize=(14,8))
plt.plot([0,1], [0,1], "k:", label="Perfectly calibrated")
plt.plot(mpv_morbid_penetrating, fop_morbid_penetrating, "s-", label="Morbidity Penetrating", color="red")
plt.plot(mpv_morbid_blunt, fop_morbid_blunt, "s-", label="Morbidity Blunt", color="firebrick")
plt.plot(mpv_mortal_penetrating, fop_mortal_penetrating, "s-", label="Mortality Penetrating", color="blue")
plt.plot(mpv_mortal_blunt, fop_mortal_blunt, "s-", label="Mortality Blunt", color="turquoise")
plt.xlabel("Mean predicted value", size=14)
plt.ylabel("Fraction of positives", size=14)
plt.legend(loc="lower right")
plt.title("Calibration plot for Morbidity and Mortality", size=22)
plt.savefig(f"{datetime.now().date()}_calibration_plots_morb_and_mort.png")
plt.show()
```

```python
labels_penetrating = pd.read_csv(data_path+"labels_penetrating.csv")
labels_blunt = pd.read_csv(data_path+"labels_blunt.csv")
for k in [1,2,3]:
    vars()[f"idx_group_{k}_penetrating"] = labels_penetrating[
        labels_penetrating.agegroup == k
    ].index.values.tolist()
    vars()[f"idx_group_{k}_blunt"] = labels_blunt[
        labels_blunt.agegroup == k
    ].index.values.tolist()

# Morbidity
y_pred_morbid_penetrating = pd.read_csv(
    data_path+"y_pred_proba_geriatric_morbid_penetrating.csv"
).iloc[:,-1]
y_pred_morbid_blunt = pd.read_csv(
    data_path+"y_pred_proba_geriatric_morbid_blunt.csv"
).iloc[:,-1]
y_test_morbid_penetrating = labels_penetrating.loc[:, "totalmorbidity"]
y_test_morbid_blunt = labels_blunt.loc[:, "totalmorbidity"]
for k in [1,2,3]:
    # Blunt
    vars()[f"y_pred_morbid_blunt_group_{k}"] = y_pred_morbid_blunt.iloc[
        vars()[f"idx_group_{k}_blunt"]
    ]
    vars()[f"y_test_morbid_blunt_group_{k}"] = y_test_morbid_blunt.iloc[
        vars()[f"idx_group_{k}_blunt"]
    ]
    # Penetrating
    vars()[f"y_pred_morbid_penetrating_group_{k}"] = y_pred_morbid_penetrating.iloc[
        vars()[f"idx_group_{k}_penetrating"]
    ]
    vars()[f"y_test_morbid_penetrating_group_{k}"] = y_test_morbid_penetrating.iloc[
        vars()[f"idx_group_{k}_penetrating"]
    ]



#Mortality
y_pred_mortal_penetrating = pd.read_csv(
    data_path+"y_pred_proba_geriatric_mortal_penetrating.csv"
).iloc[:,-1]
y_pred_mortal_blunt = pd.read_csv(
    data_path+"y_pred_proba_geriatric_mortal_blunt.csv"
).iloc[:,-1]
y_test_mortal_penetrating = labels_penetrating.loc[:, "hospitalmortality"]
y_test_mortal_blunt = labels_blunt.loc[:, "hospitalmortality"]
for k in [1,2,3]:
    # Blunt
    vars()[f"y_pred_mortal_blunt_group_{k}"] = y_pred_mortal_blunt.iloc[
        vars()[f"idx_group_{k}_blunt"]
    ]
    vars()[f"y_test_mortal_blunt_group_{k}"] = y_test_mortal_blunt.iloc[
        vars()[f"idx_group_{k}_blunt"]
    ]
    # Penetrating
    vars()[f"y_pred_mortal_penetrating_group_{k}"] = y_pred_mortal_penetrating.iloc[
        vars()[f"idx_group_{k}_penetrating"]
    ]
    vars()[f"y_test_mortal_penetrating_group_{k}"] = y_test_mortal_penetrating.iloc[
        vars()[f"idx_group_{k}_penetrating"]
    ]

print("Downloaded all the data.")
```

```python
vars()[f"fpr_morbid_blunt_group_{k}"], vars()[f"tpr_morbid_blunt_group_{k}"], vars()[f"thresh_morbid_blunt_group_{k}"] = roc_curve(
        vars()[f"y_test_morbid_bluntetrating_group_{k}"],
        vars()[f"y_pred_morbid_bluntetrating_group_{k}"]
    )
    vars()[f"auc_morbid_blunt_group_{k}"] = auc(
        vars()[f"fpr_morbid_blunt_group_{k}"],
        vars()[f"tpr_morbid_blunt_group_{k}"]
    )
```

```python
# Morbidity
for k in [1,2,3]:
    # Penetrating
    vars()[f"fpr_morbid_penetrating_group_{k}"], vars()[f"tpr_morbid_penetrating_group_{k}"], vars()[f"thresh_morbid_penetrating_group_{k}"] = roc_curve(
        vars()[f"y_test_morbid_penetrating_group_{k}"],
        vars()[f"y_pred_morbid_penetrating_group_{k}"]
    )
    vars()[f"auc_morbid_penetrating_group_{k}"] = auc(
        vars()[f"fpr_morbid_penetrating_group_{k}"],
        vars()[f"tpr_morbid_penetrating_group_{k}"]
    )
    
    # Blunt
    vars()[f"fpr_morbid_blunt_group_{k}"], vars()[f"tpr_morbid_blunt_group_{k}"], vars()[f"thresh_morbid_blunt_group_{k}"] = roc_curve(
        vars()[f"y_test_morbid_blunt_group_{k}"],
        vars()[f"y_pred_morbid_blunt_group_{k}"]
    )
    vars()[f"auc_morbid_blunt_group_{k}"] = auc(
        vars()[f"fpr_morbid_blunt_group_{k}"],
        vars()[f"tpr_morbid_blunt_group_{k}"]
    )

# Penetrating
colors_dict = {1:"g", 2:"r", 3:"b"}
plt.figure(figsize=(12, 7))
plt.title(f'ROC Curve for Geriatric Patients per Age group - Morbidity Penetrating', fontsize=16)
for k in [1,2,3]:
    plt.plot(
        vars()[f"fpr_morbid_penetrating_group_{k}"],
        vars()[f"tpr_morbid_penetrating_group_{k}"], 
        'b', label=f'Test AUC Penetrating Group {k} = {round(vars()[f"auc_morbid_penetrating_group_{k}"],3)}',
        color=colors_dict[k]
    )
plt.plot([0, 1], [0, 1], 'k--', label="Baseline AUC = 0.500")
plt.legend(loc='lower right', fontsize=12)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate', fontsize=12)
plt.xlabel('False Positive Rate', fontsize=12)
plt.savefig(f"{datetime.now().date()}_roc_curve_morbidity_penetrating_geriatric.png")
plt.show()

# Blunt
plt.figure(figsize=(12, 7))
plt.title(f'ROC Curve for Geriatric Patients per Age group - Morbidity Blunt', fontsize=16)
for k in [1,2,3]:
    plt.plot(
        vars()[f"fpr_morbid_blunt_group_{k}"],
        vars()[f"tpr_morbid_blunt_group_{k}"], 
        'b', label=f'Test AUC Blunt Group {k} = {round(vars()[f"auc_morbid_blunt_group_{k}"],3)}',
        color=colors_dict[k]
    )
plt.plot([0, 1], [0, 1], 'k--', label="Baseline AUC = 0.500")
plt.legend(loc='lower right', fontsize=12)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate', fontsize=12)
plt.xlabel('False Positive Rate', fontsize=12)
plt.savefig(f"{datetime.now().date()}_roc_curve_morbidity_blunt_geriatric.png")
plt.show()
```

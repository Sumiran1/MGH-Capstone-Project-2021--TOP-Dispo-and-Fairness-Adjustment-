## Master Imputation Function

```julia
function impute_missing_data(train, test)
    # Dropped tmode1 and hemorrhage_ctrl_type and severity_max (29/11/2019)
    # and alcohol (09/01/2020) and acslevel (01/02/2020)
    names = [
        :age, :gender, :race1, :signsoflife, :eddisp, :sbp1, :pulse1,
        :oxysat1, :temp1, :gcstot1, :bleeding_disorder, :current_chemotherapy,
        :congestive_heart_failure, :current_smoker, :chronic_renal_failure,
        :history_cva, :diabetes, :disseminated_cancer,  :copd, :steroid,
        :cirrhosis, :history_MI, :history_pvd, :hypertension_medication,
        :method_of_injury, :Head_severity, :Face_severity, :Neck_severity, :Thorax_severity, :Abdomen_severity,
        :Spine_severity, :Upper_Extremity_severity, :Lower_Extremity_severity,
        :Pelvis_Perineum_severity, :External_severity,:sbp2,:rr2,:drug_use_disorder,:issais,:morbidity,
        :drug1,:rr1,:pulse2,:acslevel,:tmode1,:gcstot2,:oxysat2,:alcohol_use_disorder,:tmode2,
        :Acute_Kidney_Injury,:ARDS, :Cardiac_Arrest_Requiring_CPR,:Deep_Surgical_Site_Infection,
        :Deep_Vein_Thrombosis, :Organ_Space_Surgical_Site_Infection,:Pulmonary_Embolism, :Unplanned_Intubation,
        :Severe_Sepsis,
        #:severity_max, #:hemorrhage_ctrl_type, #:alcohol, #:tmode1, #:acsleve  :Unplanned_Intubation,
    ]


    inc_keys_train = train[!, :inc_key]
    train = train[!, names]
    inc_keys_test = test[!, :inc_key]
    test = test[!, names]
    n_train = nrow(train)
    X = vcat(train, test)
    n_total = nrow(X)
    k_neighbors = 200
# Fitting OptImpute on train+test set at the same time (no y variable)
    lnr_optimpute_train = IAI.ImputationLearner(method=:opt_knn, knn_k=k_neighbors, cluster=true)
    IAI.fit!(lnr_optimpute_train, X)
    X = IAI.transform(lnr_optimpute_train, X)
    train_return = X[1:n_train, :]
    train_return = hcat(train_return, inc_keys_train)
    test_return = X[n_train+1:n_total, :]
    test_return = hcat(test_return, inc_keys_test)
    
    return train_return, test_return
end
```

## Blunt

```julia
data_path_blunt = "C:/Users/15125/Desktop/CAPSTONE PROJECT/CLEANED DATA/POST PRE PROCESSING/blunt/"
blunt_train =  CSV.read(data_path_blunt*"X_train_pre_process_full_blunt.csv")
blunt_test = CSV.read(data_path_blunt*"X_test_pre_process_full_blunt.csv")
imputed_blunt_train, imputed_blunt_test= impute_missing_data(blunt_train, blunt_test)
CSV.write("C:/Users/15125/Desktop/CAPSTONE PROJECT/CLEANED DATA/POST IMPUTATION/blunt/blunt_imputed_train.csv"
    ,imputed_blunt_train)
CSV.write("C:/Users/15125/Desktop/CAPSTONE PROJECT/CLEANED DATA/POST IMPUTATION/blunt/blunt_imputed_test.csv"
    ,imputed_blunt_test)
```

```python
def tree_mod(lnr, updated_lnr, cat_mapping, ord_mapping, label_mapping, trans_dict):
    # Opening JSON file
    json_file = open(lnr,)
    
    # returns JSON object as a dictionary
    data = json.load(json_file)
    
    # updating categorical features
    print("********** Updating Categorical Features *************")
    for i in range(len(data['prb_']['data']['features']['categoric_labelmap'])):
        data['prb_']['data']['features']['categoric_labelmap'][i]['levels'] = cat_mapping[i]
    
    # updating ordinal features
    print("********** Updating Ordinal Features *************")
    for i in range(len(data['prb_']['data']['features']['ordinal_labelmap'])):
        data['prb_']['data']['features']['ordinal_labelmap'][i]['levels'] = ord_mapping[i]
        
    print("********** Updating Labels *************")
    for i in range(len(data['prb_']['data']['target']['classes']['levels'])):
        data['prb_']['data']['target']['classes']['levels'][i] = label_mapping[i] 
    
    print("********** Updating Feature Names *************")
    names = data['prb_']['data']['features']['feature_names']
    trans_names = list((pd.Series(names)).map(trans_dict))
    data['prb_']['data']['features']['feature_names'] = trans_names
    
    # rewriting json
    json_to_return = json.dumps(data, indent = 4)
    with open(updated_lnr, "w") as outfile:
        outfile.write(json_to_return)
    
    # closing JSON file
    json_file.close()
    
    return data
```
